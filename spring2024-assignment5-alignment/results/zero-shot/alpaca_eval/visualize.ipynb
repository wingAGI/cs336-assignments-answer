```python
import json
with open('model_rank_qwen3-0.6B-base_VS_qwen3-1.7B_10_disprefer.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

def print_example(index):
    example = data[index]
    markdown_text = ""
    for k, v in example.items():
        # print(k, v)
        markdown_text += f"#### {k}:\n"
        markdown_text += str(v)
        markdown_text += "\n\n"

    print(markdown_text)
```

```python
print_example(0)    # disprefer的原因：输出复读机
```

```python
print_example(1)    # 长句子，不够结构化
```
